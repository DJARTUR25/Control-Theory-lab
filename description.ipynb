# Введение

Задача о стабилизации перевёрнутного маятника является одной из классических учебных задач теории управления. Простота интерпретации параметров задачи и её нелинейность делают её подходящей для демонстрации функционирования регуляторов различных типов. Целью настоящей лабораторной работы является знакомство обучающихся с возможностями и ограничениями различных регуляторов на примере этой классической задачи, а также освоение обучающимися базовых навыков численного исследования систем управления. 

# Объект управления и его модель

## Физическая и математическая модели

В настоящей работе в качестве **объекта управления (_англ._ plant, control object, process)** предполагается физический маятник — твёрдое тело, которое может вращаться вокруг неподвижной горизонтальной оси (оси подвеса) в поле вертикально направленной силы тяжести. В качестве **управления (управляющего воздействия, _англ._ control signal, input signal, control input)** выступает внешний момент сил относительно оси подвеса. Без управления маятник имеет два положения равновесия: устойчивое нижнее и неустойчивое верхнее (перевёрнутое). Задача стабилизирующего управления состоит в обеспечении устойчивости верхнего положения равновесия, то есть в обеспечении **целевого режима (_англ._ goal mode)**, при котором положение маятника остаётся в некотором смысле близким в верхнему положению равновесия. Другой возможной задачей управления является построение подъёмного регулятора, приводящее маятник в состояние, близкое к верхнему положению равновесия.

Для построения регуляторов можно использовать следующую естественную **физическую модель (_англ._ model)** такого маятника, схематично изображённую на рис.&nbsp;1. Твёрдое тело с массой $m$ закреплено на оси подвеса $O$, вокруг которой может свободно вращаться. Центр масс тела расположен на расстоянии $l > 0$ от оси $O$, а момент инерации тела относительно этой оси равен $J$. Сила тяжести направлена перпендикулярно оси $O$ и характеризуется ускорением свободного падения $g \geqslant 0$. В системе присутствует вязкое трение, характеризующееся коэффициентом $\mu \geqslant 0$ между моментом сил трения и угловой скоростью. Относительно оси $O$ к телу может прикладываться момент сил $M(t)$, $t$&nbsp;— время. Таким образом, **конфигурация системы (_англ._ configuration of the system)** в некоторый момент времени $t$ можно полностью охарактеризовать с помощью одной угловой переменной — ориентированный угол $\varphi(t)$ между направлением, противоположным силе тяжести, и направлением от оси подвеса к центру тяжести.

Консервативная динамика такой системы может быть описана функцией Лагранжа $\mathcal L$, которая в данном случае представляет собой разность кинетической $K$ и потенциальной $\Pi$ энергий, $\mathcal L = K - \Pi$. Кинетическая энергия $K$ твёрдого тела задаётся известным выражением $K = J\dot\varphi^2/2$, а потенциальная $\Pi = \Pi_1 + \Pi_2$ складываетеся из потенциальной энергии $\Pi_1 = mgl\cos\varphi$ маятника в поле силы тяжести и потенциальной энергии $\Pi_2 = -M\varphi$ взаимодействия твёрдого тела с актуатором (исполнительным устройством регулятора). Здесь и далее точка над символом обозначает полную производную по времени, $\dot\varphi = \D\varphi/\D t$. Таким образом,
$$
\mathcal L = \frac{J\dot\varphi^2}{2} - mgl \cos\varphi + M \varphi.
$$
Выражение для функции Лагранжа можно упростить до
<a name="eq:lagrangian">$$
\tag{1}
\mathcal L = J\left(\frac{\dot\varphi^2}2 - \Omega^2 \cos\varphi - u \varphi\right)
$$</a>
с помощью обозначений
$$
\Omega = \sqrt{\frac{mgl}{J}}, \quad u(t) = -\frac{M(t)}{J}.
$$

Вязкое трение в системе можно описать с помошью диссипативной функции
<a name="eq:dissipation">$$
\tag{2}
\mathcal F = \frac{\mu\dot\varphi^2}{2} = \frac{J\nu\dot\varphi^2}{2},
$$</a>
где $\nu = \mu/J$ — нормированный коэффициент вязкого трения. В данном случае диссипативная функция&nbsp;[$(2)$](#eq:dissipation) имеет смысл половинной мощности потерь (энергии, теряемой системой в единицу времени вследствие диссипации).

Уравнения Эйлера&nbsp;— Лагранжа с учётом диссипации имеют вид
<a name="eq:euler=lagrange">$$
\tag{3}
\frac{\partial \mathcal L}{\partial \varphi} - \frac{\D}{\D t} \frac{\partial \mathcal L}{\partial \dot\varphi} = \frac{\partial \mathcal F}{\partial \dot\varphi},
$$</a>
Подстановка уравнений&nbsp;[$(1)$](#eq:lagrangian) и [$(2)$](#eq:dissipative) в [$(3)$](#eq:euler-lagrange) и упрощение получившихся выражений приводит к
<a name="eq:phi">$$
\tag{4}
\ddot\varphi + \nu\dot\varphi - \Omega^2 \sin\varphi = -u.
$$</a>
Получившееся уравнение полностью описывает используемую **модель объекта управления (динамическую модель системы, _англ._ plant model, dynamical model of the system)**. В этой системе $\varphi$ и $\dot\varphi$&nbsp;— **выходные переменные (_англ._ outputs)**, $u$&nbsp;— **управление (_англ._ control input)** и $\Omega > 0$ и $\nu \geqslant 0$&nbsp;— **параметры системы (_англ._ parameters of the system)**.

Уравнение&nbsp;[$(4)$](#eq:phi) также можно переписать для векторной переменной
$$
\mathbf x \equiv \begin{bmatrix}
x_0 \\
x_1
\end{bmatrix} \equiv \begin{bmatrix}
\varphi \\
\dot\varphi
\end{bmatrix}
$$
в нормальной форме Коши как
<a name="eq:matrix">$$
\tag{5}
\dot{\mathbf x} = \mathbf f(\mathbf x, u) \equiv \begin{bmatrix}
x_1 \\
\Omega^2 \sin x_0 - \nu x_1 - u
\end{bmatrix}.
$$</a>
Нормальную форму удобно использовать при численном интегрировании системы.

## Неустойчивость верхнего положения равновесия

Система с нулевым управлением $u(t) \equiv 0$, является автономной системой, заданной на двумерном **фазовом пространстве (_англ._ phase space)** $(\varphi, \dot\varphi)$. В теории управления фазовое пространство часто также называют **пространством состояний (_англ._ state space)**. Разомкнутая система симметрична относительно замены $\varphi \rightarrow \varphi + 2\pi n$, $n$&nbsp;— целое. Эта симметрия связана с тем, что углы $\varphi$, отличающиеся на величину, кратную $2\pi$, в реальной системе отвечают одному и тому же положению маятника. Благодаря этой симметрии всё фазовое пространство (пространство состояний) можно редуцировать до фазового цилиндра, на котором $-\pi < \varphi \leqslant \pi$. На этом фазовом цилиндре разомкнутая система имеет два состояний равновесия: $\varphi = 0$, $\dot \varphi = 0$ соответствуют покоящемуся маятнику с центром тяжести в верхней точке (верхнем положении), а состояние $\varphi = \pi$, $\dot \varphi = 0$ отвечают покоящемуся маятнику с центром тяжести в нижнем положении. Линеаризация системы&nbsp;[$(5)$](#eq:matrix) вблизи верхнего равновесного состояния
$$
\mathbf x = \mathbf x_0 \equiv \begin{bmatrix}
0 \\
0
\end{bmatrix}
$$
даёт линейную систему обыкновенных дифференциальных уравнений
$$\mathbf{\delta \dot x} = \mathbf A\mathbf{\delta x},$$
где $\mathbf{\delta x} = \mathbf x - \mathbf x_0$&nbsp;— малое отклонение от равновесного состояния и $\mathbf A$&nbsp;— матрица системы,
$$\mathbf A = \begin{bmatrix}
0 & 1 \\
\Omega^2 & -\nu
\end{bmatrix}.$$
Характеристический полином матрицы $\mathbf A$ имеет вид
$$
P(\lambda) = \lambda^2 + \nu \lambda - \Omega^2.
$$
Один из двух корней $\lambda_{1, 2} = -\nu/2 \pm \sqrt{\Omega^2 - \nu^2/4}$ характеристического полинома всегда имеет положительную действительную часть. Таким образом, верхнее положения равновесия всегда неустойчиво: малые отклонения от верхнего положения равновесия со временем нарастают и груз опускается вниз, стремясь к устойчевому нижнему положению равновесия, что иллюстрирует [видео&nbsp;1](#video-1).

## Стабилизация по состоянию<a name="sec2.3"/>

Для стабилизации верхнего положения равновесия можно применить один из наиболее общих принципов построения систем управления&nbsp;— принцип **обратной связи (_англ._ feedback, feedback signal)**, который предполагает формирование управляющего воздействия на основе измеряемых или оцениваемых отклонений выходных переменных от целевых значений. При реализации принципа обратной связи **измерительное устройство (_англ._ sensor)** определяет некоторые данные о состоянии системы, которые затем поступают на вход **управляющего устройства (_англ._ controller)**, вырабатывающего управляющее воздействие, реализуемое с помощью **исполнительного устройства (_англ._ actuator)**. Объект управления, измерительное, управляющее и исполнительное устройства вместе формируют **замкнутую систему управления с обратной связью (_англ._ closed-loop feedback control system)**.

Сначала в данной работе будет рассматриваться наиболее простая ситуация, когда измерительное, управляющее и исполнительные устройства позволяют реализовать **обратную связь по состоянию (_англ._ state feedback)** с управляющим воздействием, определяемым текущим состоянием маятника $(\varphi, \dot\varphi)$. Это, в частности, предполагает мгновенную реакцию этих устройств без временных лагов (задержек). На практике это означает, что характерные времена, требуемые для измерения состояния и формирования управляющего воздействия, значительно меньше, чем собственные времена объекта управления, то есть в данной задаче значительно меньше, чем период $2\pi/\Omega$ малых колебаний и характерное время затухания колебаний $1/\nu$. Таким образом, **стратегия управления (_англ._ control strategy, control policy)** определяется функцией
<a name="eq:control">$$
\tag{6}
u = U(\varphi, \dot\varphi).
$$</a>
При таком выборе стратегии управления математическая модель замкнутой системы с обратной связью, также иногда называемой **замкнутой системой (_англ._ closed-loop system)**, получается при подстановке управления&nbsp;[$(6)$](#eq:control) в уравнение&nbsp;[$(4)$](#eq:phi),
$$
\ddot\varphi + \nu\dot\varphi - \Omega^2 \sin\varphi =  -U(\varphi,\dot\varphi),
$$
Эта модель является автономной системой, заданной на двумерном фазовом пространстве, также как и исходная **разомкнутая система (_англ._ open-loop system)** с нулевым управлением.

Так как обычно не важно, сколько оборотов совершит маятник, прежде чем придёт к вертикальному положению, можно говорить о том, что заданный целевой режим регулирования не различает углы, отличающиеся на величину, кратную $2\pi$. Соответственно при построении регулятора также не имеет смысла различать такие углы и следует выбирать стратегии управления $U(\varphi, \dot\varphi)$, периодические по $\varphi$ с периодом $2\pi$, $U(\varphi + 2\pi, \dot\varphi) = U(\varphi, \dot\varphi)$. Такие периодические стратегии управления не разрушают симметрии исходной системы относительно замены $\varphi \mapsto \varphi + 2\pi$, так что эта симметрия сохраняется и в замкнутой системе, а в качестве пространства состояний можно рассматривать фазовый цилиндр с $-\pi < \varphi \leqslant \pi$ и задавать $U$ только для таких $\varphi$. Если однако $U$ не является периодической функцией $\varphi$, то симметрия теряется и система становится чувствительной к количеству совершённых оборотов маятника. Конкретные стратегии управления, рассматриваемые в данной работе, приведены в соответствующих блоках заданий.

Линеаризованная около верхнего положения равновесия система в нормальной форме записывается как
$$\dot{\mathbf x} = \mathbf A\mathbf x + \mathbf B u,$$
где
$$
\mathbf x \equiv \begin{bmatrix}
x_0 \\
x_1
\end{bmatrix} \equiv \begin{bmatrix}
\varphi \\
\dot\varphi
\end{bmatrix}, \quad \mathbf A = \begin{bmatrix}
0 & 1 \\
\Omega^2 & -\nu
\end{bmatrix}, \quad \mathbf B = \begin{bmatrix}
0 \\
-1
\end{bmatrix}.
$$
При линейном управлении&nbsp;[$(6)$](#eq:control) (если функция $U$ дифференцируема в окрестности состояния равновесия $\varphi = 0$, $\dot\varphi = 0$) можно записать $u = \mathbf k^{\mathrm{T}} \mathbf x$, где
$$\mathbf k = \begin{bmatrix}
k_0 \\
k_1
\end{bmatrix},$$
где $k_0 = \left.(\partial U/\partial \varphi)\right|_{\substack{\varphi = 0\\\dot\varphi = 0}}$ и $k_1 = \left.(\partial U/\partial \dot\varphi)\right|_{\substack{\varphi = 0\\\dot\varphi = 0}}$&nbsp;— постоянные коэффициенты.

## Стабилизация по выходу

Описанная в подразделе [2.3](#sec2.3) обратная связь по состоянию возможна не всегда. Простейший пример этого связан с отсутствием полной информации о текущем фазовом состоянии. Для рассматриваемого маятника этот случай не столь актуален, так как две компоненты фазового состояния являются друг для друга временной производной и первообразной, а на практике производная сигнала обычно может быть востановлена по самому сигналу.

### Задержка в исполнительном устройстве<a name="sec2.4.1"/>

Более интересные ситуации возникают вследствие несовершенства измерительных и исполнительных устройств. Например, исполнительному устройству может требоваться время для формирования управляющего воздействия, которое не может устанавливаться мгновенно. В этом случае удобно разделить идеальное управляющее воздействие $u$ и реальное (действующее) управляющее воздействие $v$. Простейший для исследования вариант связи с задержкой между идеальным и реальным управляющими воздействиями задаётся уравнением первого порядка
<a name="eq:v">$$
\tag{7}
\tau_{\mathrm{a}} \dot v + v = u,
$$</a>
где $\tau_{\mathrm{a}}$&nbsp;— характерный временная задержка (лаг) между реальным и идеальным управляющим сигналом. При этом в уравнение&nbsp;[$(4)$](#eq:phi) вместо $u$ входит $v$,
<a name="eq:phi-v">$$
\tag{8}
\ddot\varphi + \nu\dot\varphi - \Omega^2 \sin\varphi = -v.
$$</a>
Теперь динамическая модель системы описывается уравнениями&nbsp;[$(7)$](#eq:v) и [$(8)$](#eq:phi-v) для фазового состояния
$$
\mathbf x \equiv \begin{bmatrix}
x_0 \\
x_1 \\
x_2
\end{bmatrix} \equiv \begin{bmatrix}
\varphi \\
\dot\varphi \\
v
\end{bmatrix}
$$
в трёхмерном фазовом пространстве.
При этом вектор
$$
\mathbf y \equiv \begin{bmatrix}
y_0 \\
y_1
\end{bmatrix} = \begin{bmatrix}
x_0 \\
x_1
\end{bmatrix} \equiv \begin{bmatrix}
\varphi \\
\dot\varphi
\end{bmatrix}
$$
выходных переменных оказывается двумерным. Отметим, что система&nbsp;[$(7)$](#eq:v) и [$(8)$](#eq:phi-v) при $\tau_{\mathrm{a}} = 0$ переходит в уравнение&nbsp;[$(4)$](#eq:phi). Нормальная форма системы&nbsp;[$(7)$](#eq:v) и [$(8)$](#eq:phi-v) имеет вид
$$
\dot{\mathbf x} = \mathbf f(\mathbf x, u) \equiv \begin{bmatrix}
x_1 \\
\Omega^2 \sin x_0 - \nu x_1 - x_2 \\
(u - x_2)/\tau_{\mathrm{a}}
\end{bmatrix},
$$
$$
\mathbf y = \begin{bmatrix}
x_0 \\
x_1
\end{bmatrix}.
$$
Стратегия управления&nbsp;[$(6)$](#eq:control) замыкает систему.

### Задержка в измерительном устройстве

Аналогично, задержка может присутствовать и в измерительном устройстве. Например, вместо $\varphi$ и $\dot\varphi$ могут быть доступны оценки $\theta$ и $\dot\theta$, где $\theta$ удовлетворяет уравнению
<a name="eq:theta">$$
\tag{9}
\tau_{\mathrm{s}} \dot \theta + \theta = \varphi
$$</a>
с характерной задержкой $\tau_{\mathrm{s}}$ между измеренным и реальным значением угла отклонения маятника. Уравнение&nbsp;[$(4)$](description.ipynb#eq:phi) совместно с [$(9)$](#eq:theta) задают динамическую модель системы для трёхмерного фазового состояния
$$
\mathbf x \equiv \begin{bmatrix}
x_0 \\
x_1 \\
x_2
\end{bmatrix} \equiv \begin{bmatrix}
\varphi \\
\dot\varphi \\
\theta
\end{bmatrix}.
$$
В этом случае вектор выходных переменных двумерен:
$$
\mathbf y \equiv \begin{bmatrix}
y_0 \\
y_1
\end{bmatrix} = \begin{bmatrix}
x_2 \\
(x_0 - x_2)/\tau_{\mathrm{s}}
\end{bmatrix} = \begin{bmatrix}
\theta \\
\dot\theta
\end{bmatrix}.
$$
Нормальная форма системы&nbsp;[$(4)$](#eq:phi) и [$(9)$](#eq:theta) имеет вид
$$
\dot{\mathbf x} = \mathbf f(\mathbf x, u) \equiv \begin{bmatrix}
x_1 \\
\Omega^2 \sin x_0 - \nu x_1 - u \\
(x_0 - x_2)/\tau_{\mathrm{s}}
\end{bmatrix},
$$
$$
\mathbf y = \begin{bmatrix}
x_2 \\
(x_0 - x_2)/\tau_{\mathrm{s}}
\end{bmatrix}.
$$
Стратегия управления при этом может иметь вид
<a name="eq:control-theta">$$
\tag{10}
u = U(\theta, \dot\theta).
$$</a>

### Задержки в измерительном и исполнительных устройствах<a name="sec2.4.3"/>

Наконец, можно рассмотреть случай, когда присутствуют задержки как в измерительном, так и в исполнительных устройствах. Динамическая модель системы при этом описывается уравнениями&nbsp;[$(7)$](#eq:v)&nbsp;— [$(9)$](#eq:theta) для четырёхмерного фазового состояния
$$
\mathbf x \equiv \begin{bmatrix}
x_0 \\
x_1 \\
x_2 \\
x_3
\end{bmatrix} \equiv \begin{bmatrix}
\varphi \\
\dot\varphi \\
\theta \\
v
\end{bmatrix}.
$$
Нормальная форма системы&nbsp;[$(7)$](#eq:v)&nbsp;— [$(9)$](#eq:theta) имеет вид
$$
\dot{\mathbf x} = \mathbf f(\mathbf x, u) \equiv \begin{bmatrix}
x_1 \\
\Omega^2 \sin x_0 - \nu x_1 - x_3 \\
(x_0 - x_2)/\tau_{\mathrm{s}} \\
(u - x_3)/\tau_{\mathrm{a}}
\end{bmatrix},
$$
$$
\mathbf y = \begin{bmatrix}
x_2 \\
(x_0 - x_2)/\tau_{\mathrm{s}}
\end{bmatrix},
$$
а стратегия управления задаётся выражением&nbsp;[$(10)$](#eq:control-theta). Линеаризованная около верхнего состояния равновесия $(\varphi, \dot\varphi, \theta, v) = (0, 0, 0, 0)$ система в нормальной форме записывается как
$$\dot{\mathbf x} = \mathbf A\mathbf x + \mathbf B u,$$
$$\mathbf y = \mathbf C\mathbf x,$$
где
$$
\mathbf x \equiv \begin{bmatrix}
x_0 \\
x_1 \\
x_2 \\
x_3
\end{bmatrix} \equiv \begin{bmatrix}
\varphi \\
\dot\varphi \\
\theta \\
v
\end{bmatrix}, \quad \mathbf y \equiv \begin{bmatrix}
y_0 \\
y_1
\end{bmatrix} \equiv \begin{bmatrix}
\theta \\
\dot\theta
\end{bmatrix},
$$
$$
\mathbf A = \begin{bmatrix}
0 & 1 & 0 & 0 \\
\Omega^2 & -\nu & 0 & -1 \\
1/\tau_{\mathrm{s}} & 0 & -1/\tau_{\mathrm{s}} & 0 \\
0 & 0 & 0 & -1/\tau_{\mathrm{a}}
\end{bmatrix}, \quad \mathbf B = \begin{bmatrix}
0 \\
0 \\
0 \\
1/\tau_{\mathrm{a}}
\end{bmatrix}, \quad \mathbf C = \begin{bmatrix}
0 & 0 & 1 & 0 \\
1/\tau_{\mathrm{s}} & 0 & -1/\tau_{\mathrm{s}} & 0
\end{bmatrix}.
$$
При линейном управлении&nbsp;[$(10)$](#eq:control-theta) (если функция $U$ дифференцируема в окрестности точки $\theta = 0$, $\dot\theta = 0$) можно записать $u = \mathbf k^{\mathrm{T}} \mathbf y$, где
$$\mathbf k = \begin{bmatrix}
k_0 \\
k_1
\end{bmatrix},$$
где $k_0 = \left.(\partial U/\partial \theta)\right|_{\substack{\theta = 0\\\dot\theta = 0}}$ и $k_1 = \left.(\partial U/\partial \dot\theta)\right|_{\substack{\theta = 0\\\dot\theta = 0}}$&nbsp;— постоянные коэффициенты.

## Вариант задачи с горизонтальной силой в качестве управляющего воздействия<a name="sec2.5"/>

Поучительным может быть сопоставление задач для описанной выше системы, где в качестве управляющего  воздействия выступает действующий на маятник момент сил, с аналогичными задачами для маятника, управляемого с помощью внешней горизонтальной силы. В этом случае вместо лагранжиана&nbsp;[$(1)$](#eq:lagrangian) следует использовать
$$
\mathcal L = J\left(\frac{\dot\varphi^2}2 - \Omega^2 \cos\varphi - u \sin\varphi\right),
$$
где $u(t) = -l_FF(t)/J$, $F(t)$&nbsp;— горизонтальная внешняя сила и $l_F$&nbsp;— расстояние от точки приложения силы до оси подвеса.
Уравнение&nbsp;[$(4)$](#eq:phi) переходит в
<a name="eq:phi-cos">$$
\tag{11}
\ddot\varphi + \nu\dot\varphi - \Omega^2 \sin\varphi = -u \cos\varphi,
$$</a>
а уравнение&nbsp;[$(8)$](#eq:phi-v) в
$$
\ddot\varphi + \nu\dot\varphi - \Omega^2 \sin\varphi = -v \cos\varphi
$$
с соответствующими модификациями нормальных форм.

При линеаризации около верхнего состояния равновесия эти два случая приводят к совершенно одинаковым уравнениям, а отличие проявляется только при рассмотрении нелинейных систем. Более того, может показаться, что при отсутствии задержек исполнительного и измерительного устройств задачи управления в двух случаях полностью математически эквивалентны: уравнение&nbsp;[$(4)$](#eq:phi) переходит в [$(11)$](#eq:phi-cos) при простой замене $u \mapsto u \cos\varphi$ и, соответственно, синтезируемые управления должны отличаться лишь множителем $\cos\varphi$. Однако надо помнить, что этот множитель $\cos\varphi$ уменьшается с ростом отклонения от целевого режима и даже может обращаться в $0$. Это приводит к существенным отличиям между двумя вариантами и в практическом плане, что будет обсуждаться подробнее в некоторых заданиях лабораторной работы.

Следует отметить, что уравнения вида&nbsp;[$(11)$](#eq:phi-cos) также возникают при рассмотрении задачи об управлении маятником с помощью горизонтального смещения точки подвеса. При этом в роли управляющего воздействия $u(t)$ выступает нормированное ускорение точки подвеса с обратным знаком, или, иными словами, в качестве горизонтальной силы $F(t)$ выступает сила инерции. При этом одно уравнение&nbsp;[$(11)$](#eq:phi-cos) само по себе позволяет лишь рассматривать задачи о стабилизации угла отклонения маятника от вертикали, а для того чтобы управлять также и горизонтальной позицией маятника необходимо дополнить систему уравнением динамики
<a name="eq:xi">$$
\tag{12}
\ddot\xi = u
$$</a>
для тела подвеса (тела, к которому подвешен маятник), здесь $\xi$&nbsp;— нормированная на длину маятника горизонтальная координата точки подвеса. Важно понимать, что такое рассмотрение пренебрегает силами реакции, действующими со стороны маятника на тело подвеса, к которому прикладывается внешняя сила. Это имеет смысл только, если масса этого тела значительно превышает массу самого маятника. В случае же сопоставимых или меньших масс, вместо уравнений&nbsp;[$(11)$](#eq:phi-cos) и [$(12)$](#eq:xi) возникнут более сложные уравнения
$$
\left(1 - \alpha\cos^2\varphi\right) \ddot\varphi + \alpha\dot\varphi^2 \sin\varphi \cos\varphi + \nu\dot\varphi - \Omega^2 \sin\varphi = -u \cos\varphi,
$$
$$
\left(1 - \alpha\cos^2\varphi\right) \ddot\xi = \alpha\dot\varphi^2 \sin\varphi + \alpha\nu\dot\varphi \cos\varphi - \alpha\Omega^2 \sin\varphi \cos\varphi + u,
$$
где $u(t)$&nbsp;— нормированная горизонтальная сила, прикладываемая к телу подвеса, и $\alpha$&nbsp;— отношение массы маятника к общей массе тела и маятника. Только в пределе бесконечно лёгких маятников, при $\alpha \to 0$, это уравнение переходит в&nbsp;[$(11)$](#eq:phi-cos). А если в системе кроме трения в оси подвеса присутствует существенное вязкое трение о внешнюю среду (например, трение тела подвеса о направляющие опоры или маятника об окружающий воздух), то в уравнении&nbsp;[$(11)$](#eq:phi-cos) возникнут слагаемые пропорциональные горизонтальной скорости тела подвеса и ограничиться рассмотрением системы второго порядка даже для стабилизации одного только угла отклонения невозможно (даже в пределе $\alpha \to 0$).
